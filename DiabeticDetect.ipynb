```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#data acquisition
df=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv")
```

```python
#bifuracation of data 
y=df["diabetes"]
X=df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
#train test split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, train_size=0.7,random_state=2529)
```

```python
#selection of model-logistic regression
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(max_iter=500)
```

```python
#train the model
model1.fit(X_train, y_train)
```

```python
#predict 
y_pred=model1.predict(X_test)
```

```python
#checking acuracy of the model
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#checking for another model-KNeighbourClassifier
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
```

```python
#train the model
model2.fit(X_train,y_train)
```

```python
#predict
y_pred=model2.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
#checking for another model-Random forest classifier
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier()
```

```python
#train the model
model3.fit(X_test,y_test)
```

```python
#predict 
y3_pred=model3.predict(X_test)
```

```python
accuracy_score(y_test,y3_pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y3_pred)
```

```python
print(classification_report(y_test,y3_pred))
```

```python
#conclusion - out of all the classification model tested, random forest 
#classifier did the best job, with accuracy as 100%.
```
